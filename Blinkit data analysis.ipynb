### Cleaning

### Business Requirements

#### KPI's

```python
# Total Sales
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# Number of Items sold
no_of_items_sold = df['Sales'].count()

# Average rating
avg_rating = df['Rating'].mean()

# printing results
print("Total Sales :", total_sales)
print("Average Sales :", avg_sales)
print("No of Items Sold :", no_of_items_sold)
print("Average Rating :", avg_rating)
```

```python
print(f"Total Sales : $ {total_sales:,.0f}")
print(f"Average Sales : $ {avg_sales:,.2f}")
print(f"No of Items Sold : {no_of_items_sold:,.0f}")
print(f"Average Rating : {avg_rating:,.2f}")
```

### Charts

#### Total Sales by Fat Content

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels = sales_by_fat.index,
                       autopct = '%.1f%%',
                        startangle = 90
       )
plt.title('Sales by Fat Content')
plt.show()
```

#### Total sales by Item Type

```python
sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)

plt.figure(figsize = (12,8))
bar = plt.bar(sales_by_item_type.index,sales_by_item_type.values)

plt.xticks(rotation = 90)
plt.xlabel("Item Type")
plt.ylabel("Sales")
plt.title("Sales by Item Type")

for bar in bar:
    plt.text(bar.get_x() + bar.get_width() / 2 , bar.get_height() ,
                            f'{bar.get_height():,.0f}', ha = 'center' , va = 'bottom',fontsize = 8)

plt.tight_layout()
plt.show()
```

#### Fat Content by Outlet for Total Sales

```python
cc = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
cc = cc[['Regular','Low Fat']] 

pl = cc.plot(kind = 'bar' , figsize = (8,5))

plt.title('Fat Content by Outlet for Sales')
plt.tight_layout()
plt.show()
```

#### Total Sales by Outlet Establishment

```python
sales_outlet = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
#sales_outlet

sales_outlet.plot(kind = 'line', marker = 'o', linestyle = '--')

#plt.plot(sales_outlet.index,sales_outlet.values,marker = 'o', linestyle = '.')

for x,y in zip (sales_outlet.index,sales_outlet.values):
    plt.text(x,y,f'{y:,.0f}',ha = 'center' , va = 'bottom',fontsize = 6)

plt.title('Total Sales by Outlet Establishment')
plt.tight_layout()
```

#### Sales by Outlet Size

```python
sales_outlet_size = df.groupby('Outlet Size')['Sales'].sum()#.sort_index(ascending = False)
sales_outlet_size

plt.figure(figsize = (6,6))
plt.title("Sales by Outlet Size")
pt = plt.pie(sales_outlet_size, startangle = 90 , 
                            autopct = '%1.1f%%', labels = sales_outlet_size.index )


#plt.legend(sales_outlet_size.index)
plt.tight_layout()
plt.show()
```

#### Sales by Outlet Location

```python
sales_outlet_loc = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_outlet_loc

plt.figure(figsize = (8,4))
plt.title("Sales by Outlet Location")

pl = sns.barplot(x = 'Sales', y = 'Outlet Location Type' , data = sales_outlet_loc)
plt.tight_layout()
plt.show()
```
